# **Pandas**

Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the python programming language.

**Pandas Series**

A pandas Series is like a column in a table. It is a 1-D array holding data of any type.

```python
import numpy as np
import pandas as pd
```

*Series from list*

```python
# string
country = ['India','USA','Nepal','Bhutan']

pd.Series(country)
```

```python
# integer
runs = [13,45,56,23,15]

pd.Series(runs)
```

```python
# custom index
marks = [67,57,89,37]
subjects = ['maths','english','science','hindi']


pd.Series(marks,index=subjects)
```

```python
# setting a name
pd.Series(marks,index=subjects,name='piyush ke marks')
```

*Series from dict*

```python
marks = {
    'maths':67,
    'english':57,
    'science':89,
    'hindi':37
}

marks_series = pd.Series(marks)
marks_series
```

*Series Attributes*

```python
# size
marks_series.size
```

```python
# dtype
marks_series.dtype
```

```python
# is_unique
marks_series.is_unique

# pd.Series([1,2,3,4,5,1]).is_unique
```

```python
# index
marks_series.index
```

```python
# values
marks_series.values
```

*Series using read_csv*
